```python
curr_dir = '/user_data/vayzenbe/GitHub_Repos/goodale-milner-test'

import sys
sys.path.insert(0,curr_dir)
import os
import skimage
import numpy as np
import matplotlib.pyplot as plt


from glob import glob as glob
```

```python
'''
Resize stim
'''

stim_dir = f'{curr_dir}/stim/og'
out_dir = f'{curr_dir}/stim/all'
os.makedirs(out_dir,exist_ok=True)

stim_files = glob(f'{stim_dir}/*.jpg')
rots = np.arange(0,360,15)

#propotions in steps of .2
props = np.arange(.5,1.25,.25)

#loop through images
for stim_file in stim_files:
    #create folder for each image
    stim_name = stim_file.split('/')[-1].split('.')[0]

    #create folder
    os.makedirs(f'{out_dir}/{stim_name}', exist_ok=True)


    #load image
    stim = skimage.io.imread(stim_file)

    #grayscale image
    if len(stim.shape) > 2:
        stim = skimage.color.rgb2gray(stim)

    #find midpoint of image
    mid = np.array(stim.shape)/2

    #loop through rotations
    for size in props:

        #rescale image
        stim_rescale = skimage.transform.rescale(stim,size)

        #pad rescaled image so it is the same size as the original
        stim_pad = np.pad(stim_rescale,((int(mid[0]-stim_rescale.shape[0]/2),int(mid[0]-stim_rescale.shape[0]/2)),(int(mid[1]-stim_rescale.shape[1]/2),int(mid[1]-stim_rescale.shape[1]/2))),'constant')

              


        #save image
        skimage.io.imsave(f'{out_dir}/{stim_name}/{stim_name}_{size}.jpg',stim_pad)



    
```

```python
'''
Rotate stim
'''


stim_dir = f'{curr_dir}/stim/og'
out_dir = f'{curr_dir}/stim/orient'

stim_files = glob(f'{stim_dir}/*.jpg')

rots = np.arange(0,360,20)

#loop through images
for stim_file in stim_files:
    #create folder for each image
    stim_name = stim_file.split('/')[-1].split('.')[0]

    #create folder
    os.makedirs(f'{out_dir}/{stim_name}', exist_ok=True)


    #load image
    stim = skimage.io.imread(stim_file)

    #grayscale image
    if len(stim.shape) > 2:
        stim = skimage.color.rgb2gray(stim)


    #loop through rotations
    for rot in rots:
        #rotate image
        stim_rot = skimage.transform.rotate(stim,rot)

        #save image
        skimage.io.imsave(f'{out_dir}/{stim_name}/{stim_name}_{rot}.jpg',stim_rot)


    
```
